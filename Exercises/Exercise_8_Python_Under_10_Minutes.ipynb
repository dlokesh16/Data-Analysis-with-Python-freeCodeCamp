### Exercise 4

Make a function that receives a list as parameter and return how many elements it has; if it doesn't have any element return an error.

```python
# your code goes here
def myfunction(*args):
    size = len(args)
    
    if(size > 0):
        return 'List has {} elements'.format(size)
    else:
        return 'List has no elements'
```

```python
def list_size(my_list):
    size = len(my_list)
    
    if (size > 0):
        return 'List has {} elements'.format(size)
    else:
        return 'Error, no elements'
```

```python
list_size([1, 2, 3])
```

```python
list_size([])
```

```python
myfunction([1, 2, 3])
```

```python
myfunction([])
```
